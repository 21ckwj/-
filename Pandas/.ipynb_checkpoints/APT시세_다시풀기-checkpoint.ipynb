```python
# 평당 가격을 구하기 위해 데이터타입을 바꿔줘야 함
```

```python
df['분양가격(제곱미터)'].astype('float') # 바꿔주려했으나 에러 : string 값 있음('' 이런값은 숫자로 바꿀 수 없음)
```

```python
# 모두 문자열로 바꾸고 전처리

df['분양가격(제곱미터)'] = df['분양가격(제곱미터)'].astype(str)
```

```python
sorted(df['분양가격(제곱미터)'].unique())
```

```python
# for 문을 이용해 index값, error값들을 모으자

index = []
errors= []


for i,j in enumerate(df['분양가격(제곱미터)']):
    try:
        float(j)
        index.append(i)
    except:
        errors.append(j)

errors
```
