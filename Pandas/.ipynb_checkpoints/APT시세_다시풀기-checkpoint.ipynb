```python
#결측치 확인
df.isnull().sum()
```

```python
# 1. 평당 가격 = 분양가격(제곱미터) * 3.3
# 데이터타입 확인
df.info()  # 분양가격 = object  바꿔줘야함
```

```python
# df['분양가격(제곱미터)'].astype(float) # '' 때문에 float으로 바로 변환 불가
```

```python
df['분양가격(제곱미터)'] = df['분양가격(제곱미터)'].astype(str)
```

```python
# 전처리 전 대략적 데이터 확인
list(df['분양가격(제곱미터)'].unique())
```

```python
# for문을 통한 에러값 모아주기
lndex = []
error =[]

for i, j in enumerate(df['분양가격(제곱미터)']):  # i= 인덱스 , j= 칼럼값
    try:
        float(j)
        lndex.append(i)
    except:
        error.append(j)
```

```python
# 모아진 에러값들 확인
print(error)  # 공백, 쉼표(,), -
```

```python
# 에러값들 전처리

df['분양가격(제곱미터)']= df['분양가격(제곱미터)'].replace(' ','').repalce('  ','').replace(',','').replace('-','')
df['분양가격(제곱미터)']
```
