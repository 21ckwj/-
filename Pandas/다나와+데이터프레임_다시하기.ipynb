```python
# 코롤링 데이터로 데이터프레임 만들기 위해 각각을 리스트에 담기

title_lst= []
spec_lst= []
price_lst= []

for tag in tags:
    try:
        title = tag.select('div > div.prod_info > p > a')[0].text
        spec = tag.select('div > div.prod_info > dl > dd > div')[0].text
        price = tag.select(' p.price_sect> a > strong')[0].text
        
        print('제품명: ', title)
        title_lst.append(title)
        
        print('제품성능: ', spec)
        spec_lst.append(spec)
        
        print("가격: ", price)
        price_lst.append(price)
        
        print('\n')
    except:
        pass
```

```python
# 데이터프레임 만들기

df = pd.DataFrame({'제품명': title_lst,'성능': spec_lst, '가격': price_lst})
df.head()
```

```python
# 코롤링 데이터로 데이터프레임 만들기 위해 각각을 리스트에 담기
# 글씨 깨진행의 경로가 id = 'prod_item product-pot'

title_lst= []
spec_lst= []
price_lst= []

for tag in tags:
    try:
        if tag['id']=='prod_item product-pot':
            pass
        else:
            title = tag.select('div > div.prod_info > p > a')[0].text
            spec = tag.select('div > div.prod_info > dl > dd > div')[0].text
            price = tag.select(' p.price_sect> a > strong')[0].text
        
            print('제품명: ', title)
            title_lst.append(title)
        
            print('제품설명: ', spec)
            spec_lst.append(spec)
        
            print("가격: ", price)
            price_lst.append(price)
        
            print('\n')
        
    except:
        pass
```

```python
# 데이터프레임 만들기

df = pd.DataFrame({'제품명': title_lst,'제품설명': spec_lst, '가격': price_lst})
df.head()
```
