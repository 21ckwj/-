### 크롤링 연습   
 검색어를 입력 받은 후 해당 검색어로 검색한 결과 다음 뉴스 페이지에서   
   
 검색한 결과의 제목, url, summary를 크롤링하시오

```python
# 크롤링 과정 정리

# 1. 원하는 정보가 있는 페이지 전체 url을 가져온다
# 2. requests.get(url)을 통해 페이지 정보 저장
# 3. BeautifulSoup( , ) 함수를 통해 데이터 추출 용이하게 만듬(select 함수를 사용할 수 있음)

# 4. 원하는 정보만 있는 부분의 path를 개발자 도구를 통해 가져온다

# 5. soup 에서 원하는 정보가 있는 path 부분을 찾아준다.
# 6. 원하는 부분 더 세세히 찾아 완성(.text , or ['href']..)
```

```python
#1. 검색시 url 가져오기

search_kw = input("검색어를 입력하세요 ")
url = "https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q={}".format(search_kw)
```

```python
#2. requests.get(url)을 통해 페이지 정보 저장
import requests
from bs4 import BeautifulSoup

res = requests.get(url)
res.text
```

```python
# 3. BeautifulSoup( , ) 함수를 통해 데이터 추출 용이하게 만듬(soup)

soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
# 4. 원하는 정보만 있는 부분의 path를 개발자 도구를 통해 가져온다

# path = '#newsColl > div.cont_divider > ul > li:nth-child(1)'
path = '#newsColl > div.cont_divider > ul > li' #  #=id 밑에 newsColl div class 밑에 
```

```python
# 5. soup 에서 원하는 정보가 있는 path 부분을 찾아준다.
tags = soup.select(path)
# tags # 너무 길어
tags[0]

# 제목  <a class="tit_main fn_tit_u" href="http://v.media.daum.net/v/20220119093300994?f=o" onclick='smartLog(this, "dc=NNS&amp;d=2691-z8SO9QvxEEGp_&amp;pg=1&amp;r=1&amp;p=4&amp;rc=10&amp;e1=16O8sbGdUflWYlfqu4&amp;e3=0&amp;ext=dsid=2691-z8SO9QvxEEGp_", event, {"cpid": {"value": "16O8sbGdUflWYlfqu4"}});' target="_blank"><b>블록</b><b>체인</b>으로 의약물류 .. 데이터 위변조 '멈춰!'</a>
# url  <a class="thumb" href="http://v.media.daum.net/v/20220119093300994?f=o"
# summary  
# <p class="desc">관리에 대한 고민이 깊어질 수밖에 없다. 특히 신선식품과 의약품은 유통과정에서의 안전성 확보가 필수. 국내외 물류업계가 <b>블록</b><b>체인</b> 기술 도입을 서두르는 이유다. 가상화폐 비트코인의 원천기술로 유명한 <b>블록</b><b>체인</b>은 거래 기록을 중앙 서버 한 곳에 집중시키지...</p>
```

```python
def crawling():
    search_kw = input("검색어를 입력하세요: ")
    url = "https://search.daum.net/search?nil_suggest=btn&w=news&DA=SBC&q={}".format(search_kw)
    
    import requests
    from bs4 import BeautifulSoup
    res = requests.get(url)
    
    soup = BeautifulSoup(res.text,'html.parser')
    
    path = '#newsColl > div.cont_divider > ul > li'
    
    tags = soup.select(path)
#     title = tags[0].select("a.tit_main")[0].text
#     urlf = tags[0].select("a")[0]['href']
#     summary = tags[0].select("p.desc")[0].text
    
    for t in tags:
        title = t.select("a.tit_main")[0].text
        urlf = t.select("a")[0]['href']
        summary = t.select("p.desc")[0].text
        
    print("제목: {}\nURL: {}\nsummary: {}\n".format(title,urlf,summary))

crawling()
    
```
